## 3. Exploring measures per property type and districts in Berlin (avg monthly price, avg total living space).

```python
#let's see property type offerings in Berlin
fig = px.histogram(df, 
                   x='property',
                   color='property',
                   title="Available property type in Berlin",
                   labels={
                      "property": "Property type",
                      "count": " ",
                      "property": "Property type"
                   })
fig.update_xaxes(categoryorder='total descending', tickangle = 60)
fig.show()
```

```python
#let's see how many offers each district has by property type
fig = px.histogram(df, 
                   x='district', 
                   color='property',
                   title="Available property type in Berlin by district",
                   labels={
                      "district": "District name",
                      "count": " ",
                      "property": "Property type"
                   })
fig.update_xaxes(categoryorder='total descending', tickangle = 60)
fig.show()

fig = px.histogram(df, 
                   x='district', 
                   color='property', 
                   barmode='stack',
                   barnorm='percent',
                   title="Percent share of available property type in Berlin by district",
                   labels={
                      "district": "District name",
                      "count": " ",
                      "property": "Property type"
                   })

fig.update_xaxes(categoryorder='category ascending', tickangle = 60)
fig.show()
```

```python
#lets calculate average price per property type in each district

fig = px.histogram(df, 
                   x='district',
                   y='monthly_price_avg', 
                   color="property",
                   histfunc='avg',
                   barmode='group',
                   barnorm=None,
                   title="Average monthly price per property type in Berlin",
                   labels={
                      "district": "District name",
                      "monthly_price_avg": "monthly price (EUR)",
                      "property": "Property type"
                   })
fig.update_xaxes(categoryorder='sum descending', tickangle = 60)
fig.show()

#'trace', 'category ascending', 'category descending',
#'array', 'total ascending', 'total descending', 'min ascending', 
#'min descending', 'max ascending', 'maxdescending', 
#'sum ascending', 'sum descending', 'mean ascending', 
#'mean descending', 'median ascending', 'median descending'
```

```python
#lets see avg living space per property type per each district
fig = px.histogram(df, 
                   x='district',
                   y='total_space', 
                   color="property",
                   histfunc='avg',
                   barmode='group',
                   barnorm=None,
                   title="Average living space per property type in Berlin",
                   labels={
                      "district": "District name",
                      "total_space": "total living space (m²)",
                      "property": "Property type"
                   })
fig.update_xaxes(categoryorder='max descending', tickangle = 60)
fig.show()
```

```python
#adding Long Lat to the dataframe 
df_district = df.set_index('district')
df_coordinates = pd.read_csv('df_districts_coordinates.csv', index_col='district')
df = pd.merge(df_district, df_coordinates, left_index=True, right_index=True)
df.reset_index(inplace=True)
```

```python
#let's have a look at average price for apartments per district in Berlin
df_1 = df[df['property']=='Apartment'].groupby('district').mean()[['monthly_price_avg', 'Latitude', 'Longitude']]
df_1.reset_index(inplace=True)
fig = px.scatter_mapbox(df_1,
                        text=df_1['district'],
                        lat="Latitude", 
                        lon="Longitude",
                        color="monthly_price_avg",
                        size="monthly_price_avg",
                        color_continuous_scale=px.colors.cyclical.IceFire, 
                        size_max=15, 
                        zoom=9,
                        mapbox_style="carto-positron",
                        title="Average price for apartments per district in Berlin",
                        labels={
                        "monthly_price_avg": "avg price"
                        },)
fig.show()
```

We can see that the most expensive offers are in __Kreuzberg, Tiergarden, Moabit__. We can also see an outlier - __Wilmersdorf__. Property size and small amount of apartments in this area might contribute. Let's calculate average price per square meter and compare the results.

```python
df['sqr_m_price_avg'] = df.apply(lambda x: x['monthly_price_avg'] / x['total_space'], axis=1)
```

```python
df_2 = df[df['property']=='Apartment'].groupby('district').mean()[['sqr_m_price_avg', 'Latitude', 'Longitude']]
df_2.reset_index(inplace=True)
fig = px.scatter_mapbox(df_2,
                        text=df_2['district'],
                        lat="Latitude", 
                        lon="Longitude",
                        color="sqr_m_price_avg",
                        size="sqr_m_price_avg",
                        color_continuous_scale=px.colors.cyclical.IceFire, 
                        size_max=15, 
                        zoom=9,
                        mapbox_style="carto-positron",
                        title="Average price per m² for apartments in Berlin",
                        labels={
                        "sqr_m_price_avg": "avg price per m²"
                        },)
fig.show()
```

As we can see __Mitte__ and __Prenzlauer Berg__ have the highest price for apartment per m² in Berlin - __30.3__ and __30.4 EUR__ respectively. __Wilmersdorf__ is somewhere in the middle with it's __25.8 EUR__.

## 4. Comparing features availability for each district in Berlin (furnished, washing machine, kitchen, elevator, balcony).

```python
#let's see propert offers in Berlin including different features. 

#FURNISHED property offers in Berlin
fig1 = px.histogram(df, 
             x="district",
             color="feature_furnished_yn",
             facet_col="property",
             category_orders={"feature_furnished_yn": [1, 0]},
             title="FURNISHED property offers in Berlin",
             labels={"property": "Property type",
                     "feature_furnished_yn": "Furnished"
             })
#Property offers with BALCONY in Berlin
fig2 = px.histogram(df, 
             x="district",
             color="feature_balcony_yn",
             facet_col="property",
             category_orders={"feature_balcony_yn": [1, 0]},
             title="Property offers with BALCONY in Berlin",
             labels={"property": "Property type",
                     "feature_balcony_yn": "Balcony"
             })
#Property offers with ELEVATOR in Berlin
fig3 = px.histogram(df, 
             x="district",
             color="feature_elevator_yn",
             facet_col="property",
             category_orders={"feature_elevator_yn": [1, 0]},
             title="Property offers with ELEVATOR in Berlin",
             labels={"property": "Property type",
                     "feature_elevator_yn": "Elevator"
             })

fig1.show()
fig2.show()
fig3.show()
```
